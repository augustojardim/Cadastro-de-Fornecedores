```python
#FUNÇÃO ADICIONAR
def adicionar(cadastro):
    nom_novo = input('Insira o nome do novo fornecedor: ').title()
    # Verificação se o telefone inserido é composto somente de números e se possui 11 dígitos com o ddd
    tel_novo = input('Insira o telefone do novo fornecedor com o ddd: ')
    while not (tel_novo.isdigit()) or ((len(tel_novo)!=11) and (len(tel_novo)!=10)):
        if not tel_novo.isdigit():
            print('Resposta inválida! Somente números devem ser digitados.')
            tel_novo = input('Insira o telefone do novo fornecedor: ')
        else:
            print('Resposta inválida! O número de telefone com ddd deve possuir 10 dígitos ou 11 dígitos.')
            tel_novo = input('Insira o telefone do novo fornecedor: ')
    ema_novo = input('Insira o email do novo fornecedor: ')
    # Geração do novo código automaticamente
    tamanho = len(cadastro['Código']) + 1
    cod_novo = f'{tamanho}'
    i = 0
    # Permite a geração de códigos automáticos únicos 
    while cod_novo in cadastro['Código']: 
        tamanho = len(cadastro['Código']) + i
        cod_novo = f'{tamanho}'
        i = i + 1

    # Adição das informações no cadastro
    cadastro['Código'].append(cod_novo)
    cadastro['Nome'].append(nom_novo)
    cadastro['Telefone'].append(tel_novo)
    cadastro['Email'].append(ema_novo)
    return cadastro

# FUNÇÃO BUSCAR
def buscar(cadastro):
    busca = input('Insira o código do fornecedor que deseja obter informações: ')
    while busca not in cadastro['Código']:
        print('Código não encontrado!')
        busca = input('Insira novamente o código: ')
    idx_cod = cadastro['Código'].index(busca)
    cod_mostra = cadastro['Código'][idx_cod]
    nom_mostra = cadastro['Nome'][idx_cod]
    tel_mostra = cadastro['Telefone'][idx_cod]
    ema_mostra = cadastro['Email'][idx_cod]
    print(f'As informações do fornecedor são:')
    print(f'O código é {cod_mostra}')
    print(f'O nome é {nom_mostra}')
    print(f'O telefone é {tel_mostra}')
    print(f'O email é {ema_mostra}')
    
#FUNÇÃO APAGAR
def apagar(cadastro):
    apg = input('Insira o código do fornecedor que deseja deletar do cadastro: ')
    while apg not in cadastro['Código']:
        print('Código não encontrado!')
        apg = input('Insira novamente o código: ')
    idx_cod = cadastro['Código'].index(apg)
    cod_mostra = cadastro['Código'].pop(idx_cod)
    nom_mostra = cadastro['Nome'].pop(idx_cod)
    tel_mostra = cadastro['Telefone'].pop(idx_cod)
    ema_mostra = cadastro['Email'].pop(idx_cod)
    print(f'Fornecedor {apg} deletado')
    return cadastro
```

```python
# Cadastro Inicial usado como base para verificar o funcionamento do sistema
cadastro = {'Código':[],
            'Nome':[],    
            'Telefone':[],
            'Email':[]}

#Programa Principal
print('Bem vindo ao sistema de cadastro de fornecedores')
#Flag usada para integrar as três funções e para fazer com que o programa continue rodando
#a menos que o usuário selecione a opção de sair dos sistema de cadastro
flag = True 
while flag == True:
    print('Digite a opção referente ao que você deseja fazer')
    print('Digite 1 para adicionar novos fornecedores: ')
    print('Digite 2 para apagar dados de fornecedores: ')
    print('Digite 3 para buscar dados de fornecedores: ')
    print('Digite 0 se deseja sair do sistema: ')
    opcoes = ['1','2','3','0']
    opc = input('Digite o comando desejado: ')
    # Verificando se o usuário está usando uma das opções permitidas
    while opc not in opcoes:
        print('Comando não reconhecido!')
        opc = input('Digite novamente o comando desejado: ')
    # Este bloco não permite ao usuário acessar a opção de apagar dados quando o cadastro estiver vazio
    while (not cadastro['Código']) and (not cadastro['Nome']) and (not cadastro['Telefone'])\
    and (not cadastro['Email']) and (opc == '2'):
        print('O cadastro está vazio! Você não pode usar a função apagar.')
        opc = input('Digite novamente o comando desejado: ')
    # Este bloco não permite ao usuário acessar a opção de buscar dados quando o cadastro estiver vazio
    while (not cadastro['Código']) and (not cadastro['Nome']) and (not cadastro['Telefone'])\
    and (not cadastro['Email']) and (opc == '3'):
        print('O cadastro está vazio! Você não pode usar a função buscar.')
        opc = input('Digite novamente o comando desejado: ')
#  Chama função de adicionar fornecedor        
    if opc == '1':
        adicionar(cadastro)
#  Chama função de apagar fornecedor
    elif opc == '2':
        cadastro = apagar(cadastro)
#  Chama função para buscar cadastro
    elif opc == '3':
        buscar(cadastro)
# Encerra o programa e possibilita a saída do sistema de cadastro
    elif opc == '0':
        flag = False
        print('Você saiu do sistema!')
```
